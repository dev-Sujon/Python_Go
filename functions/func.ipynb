Pass by Reference vs Value
In programming languages like C and C++, there are two main ways to pass variables to a function, which are Call by Value and Call by Reference (also known as pass by reference and pass by value). However, the way we pass variables to functions in Python differs from others.

call by value − When a variable is passed to a function while calling, the value of actual arguments is copied to the variables representing the formal arguments. Thus, any changes in formal arguments does not get reflected in the actual argument. This way of passing variable is known as call by value.

call by reference − In this way of passing variable, a reference to the object in memory is passed. Both the formal arguments and the actual arguments (variables in the calling code) refer to the same object. Hence, any changes in formal arguments does get reflected in the actual argument.

```python
## Write function to add two number
## Write a function to add two numbers.
## Write a function to subtract two numbers.
## Write a function to multiply two numbers.
## Write a function to divide two numbers
## .Write a function to find the square of a number.
## .Write a function to find the cube of a number.

class Calc:
    # Addition
    def addition(self, num1, num2):
        return num1 + num2

    # Subtraction
    def subtraction(self, num1, num2):
        return num1 - num2

    # Multiplication
    def multiplication(self, num1, num2):
        return num1 * num2

    # Division
    def division(self, num1, num2):
        if num2 == 0:
            return "Error! Division by zero."
        return num1 / num2
    # Square
    def square(self, num):
        return num**2
    ## Cube
    def Cube(self, num):
        return num**3
    
# Example usage
calculator = Calc()
print(calculator.addition(5, 3))       # 8
print(calculator.subtraction(5, 3))    # 2
print(calculator.multiplication(5, 3)) # 15
print(calculator.division(5, 0))       # Error! Division by zero.
print(calculator.square(2))
print(calculator.Cube(3))
```

```python
    #.Write a function to check if a number is even.
    #.Write a function to check if a number is odd.
    #.Write a function to find the maximum of two numbers.
    #.Write a function to find the minimum of two numbers.
    #.Write a function to check if a number is positive.
    #.Write a function to check if a number is negative.

class NumberChecker:
    # Check if a number is even or odd
    def isEvenOdd(self, num):
        if num % 2 == 0:
            return "even"
        else:
            return "odd"

    # Find maximum and minimum of two numbers
    def maxMin(self, num1, num2):
        if num1 > num2:
            return f"{num1} is Max, {num2} is Min"
        else:
            return f"{num2} is Max, {num1} is Min"

    # Check if a number is positive or negative
    def isPositiveNegative(self, num):
        if num > 0:
            return "positive"
        elif num < 0:
            return "negative"
        else:
            return "zero"

# Example usage
checker = NumberChecker()
print(checker.isEvenOdd(5))         # odd
print(checker.maxMin(10, 7))        # 10 is Max, 7 is Min
print(checker.isPositiveNegative(-3)) # negative
```

```python
#Write a function to convert Celsius to Fahrenheit.

#Write a function to convert Fahrenheit to Celsius.

#Write a function to calculate the factorial of a number.

#Write a function to calculate the sum of first n natural numbers.

#Write a function to calculate the product of first n natural numbers.

#Write a function to find the remainder of two numbers.

#Write a function to check if a number is divisible by another.

#Write a function to swap two numbers.

class MathOperations:
    # Convert Celsius to Fahrenheit
    def celsiusToFahrenheit(self, celsius):
        return (celsius * 9/5) + 32

    # Convert Fahrenheit to Celsius
    def fahrenheitToCelsius(self, fahrenheit):
        return (fahrenheit - 32) * 5/9

    # Calculate factorial of a number
    def factorial(self, num):
        if num < 0:
            return "Undefined for negative numbers"
        fact = 1
        for i in range(1, num + 1):
            fact *= i
        return fact

    # Sum of first n natural numbers
    def sumNaturalNumbers(self, n):
        return n * (n + 1) // 2

    # Product of first n natural numbers
    def productNaturalNumbers(self, n):
        product = 1
        for i in range(1, n + 1):
            product *= i
        return product

    # Find remainder of two numbers
    def remainder(self, num1, num2):
        if num2 == 0:
            return "Division by zero error"
        return num1 % num2

    # Check if a number is divisible by another
    def isDivisible(self, num1, num2):
        if num2 == 0:
            return False
        return num1 % num2 == 0

    # Swap two numbers
    def swap(self, num1, num2):
        return num2, num1

# Example usage
math_ops = MathOperations()
print(math_ops.celsiusToFahrenheit(0))       # 32.0
print(math_ops.fahrenheitToCelsius(100))     # 37.777...
print(math_ops.factorial(5))                 # 120
print(math_ops.sumNaturalNumbers(10))        # 55
print(math_ops.productNaturalNumbers(4))     # 120
print(math_ops.remainder(10, 3))             # 1
print(math_ops.isDivisible(10, 3))           # False
print(math_ops.swap(5, 9))                   # (9, 5)
```
